## Jobs Analysis using Python and SQL


Simple data analysis using Python and Sql from an American Community Survey on job outcome statistics based on college majors

Practicing Basic SQL for  #66daysofdata

```python
import os
import sqlite3

DEFAULT_PATH = os.path.join(os.getcwd(),'~/db/jobs.db')

def db_connect(db_path=DEFAULT_PATH):
    connection = sqlite3.connect(db_path)
    cursor = connection.cursor()
    return cursor
```

```python
DEFAULT_PATH
```

```python
cur = db_connect(DEFAULT_PATH)
```

```python
schema = "SELECT name \
    FROM sqlite_master \
    WHERE type = 'table';"
cur.execute(schema).fetchall()
```

```python
pragma = "PRAGMA table_info ('recent_grads')"
cur.execute(pragma).fetchall()
```

```python
all_grads = "SELECT * \
    FROM recent_grads"
cur.execute(all_grads).fetchall()
```

```python
limit_grads = "SELECT * \
    FROM recent_grads \
    LIMIT 5"
cur.execute(limit_grads).fetchall()
```

```python
major_and_sharewomen = "SELECT Major, ShareWomen \
    FROM recent_grads \
    LIMIT 5"
cur.execute(major_and_sharewomen).fetchall()
```

```python
major_and_sharewomen = "SELECT Major, ShareWomen \
    FROM recent_grads \
    WHERE ShareWomen < 0.5"
cur.execute(major_and_sharewomen).fetchall()
```

```python
eng_women = "SELECT Major \
    FROM recent_grads \
    WHERE Major_category = 'Engineering' \
    AND ShareWomen > 0.5"
cur.execute(eng_women).fetchall()
```

```python
employed_women_majors = "SELECT Major, ShareWomen, Unemployment_rate \
    FROM recent_grads \
    WHERE ShareWomen > 0.3 AND Unemployment_rate < 0.1 \
    ORDER BY Unemployment_rate DESC"
cur.execute(employed_women_majors).fetchall()
```
