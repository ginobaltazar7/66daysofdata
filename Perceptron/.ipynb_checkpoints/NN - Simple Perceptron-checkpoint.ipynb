```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# types of activation functions

def activat1(x):
    return np.tanh(x)         # tanh or f(x) = 1 — exp(-2x) / 1 + exp(-2x);

def activat2(x):              # stochastically determined +/- binary activation
    if x >= 0:
        return 1
    else:
        return -1

def activat3(x):             # sigmoid -> f(x) = 1 / 1 + exp(-x)
    return 1 / (1 + np.exp(-x))
```

```python
#perceptron class using activate function 2
class perceptron2(object):
    def __init__(self,dim):
        self.dim = dim       #input space dimension
        self.learner = 0.1   #learning rate
        self.w = np.array([0.001]*(dim+1)) #+1 for bias
        self.cycles = 0      #counter for number of loop cycles
        self.converger = 100 #check for convergence
        self.maxiter = 10000 #max num of iterations before the loop stops
        self.toler = 0.001   #tolerance
        self.errors = 0      #number of wrong predictions in the testing set 
    
    #training algorithm
    def train(self, X, y):
        
        little_w = 0                               #counter for little changes of bias
        
        while(little_w < self.converger and self.cycles < self.maxiter):
            
            i = np.random.randint(0, len(X))
            x_vec  = np.concatenate(([1],X[i]))    #random biased input vector
            d_output = y[i]                        #desired output for the selected example
            a_output = np.dot(self.w, x_vec)       #computed actual output 
            
            old_w = self.w
            act = activat2(a_output)               #apply activation to de-linearize actual output 
            self.w += self.learner*(d_output - act)*x_vec #Learning Rule = compare desired output and actual output

            if np.linalg.norm(self.w - old_w) < self.toler:
                little_w += 1
            else:
                little_w = 0                       #if w changed enough, reset the counter

            self.cycles += 1
    
    #predict algorithm        
    def predict(self, X_pred, y_des):
        y_pred = ['error']*len(X_pred)                  #initialize the vector of predictions
        for i in range(len(X_pred)):
            x_vec = np.concatenate(([1],X_pred[i]))     #biased input vector 
            p_output = np.dot(self.w, x_vec)            #compute predicted output
            y_pred[i] = activat2(p_output)              #apply activation to de-linearize computed prediction
            if (y_pred[i] != y_des[i]):                 #if the prediction is incorrect, increment errors counter
                self.errors +=  1       
            
        return y_pred
```

```python
# Train Set
Npoints = 100
xcomp = np.linspace(2,np.pi, Npoints)

redRdm = np.random.normal(size=Npoints, loc=0, scale=2) # random sample from a normal distribution
redPts = np.array([np.array([xcomp[i],3*xcomp[i] + redRdm[i]]) for i in range(100)])
redDes = np.repeat(-1,Npoints)

blueRdm = np.random.normal(size=Npoints, loc=8, scale=2)
bluePts = np.array([ np.array([xcomp[i],3*xcomp[i] + blueRdm[i]]) for i in range(100)])
blueDes = np.repeat(1,Npoints)

X_train = np.concatenate((redPts,bluePts))
y_train = np.concatenate((redDes,blueDes))
```

```python
# Test Set
Npoints = 50
xcomp = np.linspace(0, np.pi, Npoints)

redRdm = np.random.normal(size=Npoints, loc=0, scale=2)
redPts = np.array([np.array([xcomp[i],3*xcomp[i] + redRdm[i]]) for i in range(Npoints)])
redDes = np.repeat(-1,Npoints)

blueRdm = np.random.normal(size=Npoints, loc=8, scale=2)
bluePts = np.array([ np.array([xcomp[i],3*xcomp[i] + blueRdm[i]]) for i in range(Npoints)])
blueDes = np.repeat(1,Npoints)

X_test = np.concatenate((redPts,bluePts))
y_test = np.concatenate((redDes,blueDes))
```

```python
# Model using activate function 2
dimension = X_train.shape[1]
Blueberrytron = perceptron2(dimension)
Blueberrytron.train(X_train, y_train)
Blueberrytron.predict(X_test, y_test)
```

```python
plt.figure(figsize=(15,10))
for i in range(Npoints):
    plt.scatter(redPts[i,0],redPts[i,1],alpha = 0.3,c='r')
    plt.scatter(bluePts[i,0],bluePts[i,1],alpha = 0.3,c='b')
weights = Blueberrytron.w
plt.plot(xcomp, -(weights[1]*xcomp + weights[0])/weights[2] , 'g-',label='Learned Classifier Line') 
plt.savefig('Blueberrytron.png')
plt.title('Perceptron')
plt.legend()
plt.show()
```

# Observations

The behavior of the perceptron's simple learning function converges well to a classifier line however with the sigmoid activation function, as written, the learning degrades because of the gradient problem. That is, a continuous activation function needs to be continuously differentiable so the steepest descent can be fully utilized while minimizing the error function. Increasing the iterations ten-fold (epochs) helped compensate; rewriting the sigmoid activation to find the best gradient is a good next step.
